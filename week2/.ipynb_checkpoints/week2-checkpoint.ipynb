```python
# 1. Create a NumPy array of zeros with shape (3, 4)
zeros_array = np.zeros((3, 4))
print("Zeros Array:")
print(zeros_array)
```

```python
# 3. Create a 5x5 identity matrix and extract the diagonal elements into a 1D array
identity_matrix = np.eye(5)
diagonal_elements = np.diag(identity_matrix)
print("\nIdentity Matrix:")
print(identity_matrix)
print("\nDiagonal Elements:")
print(diagonal_elements)
```

```python
# 4. Create a 4x5 array of even numbers starting from 10
even_matrix = np.arange(10, 50, 2).reshape(4, 5)
print("\n4x5 Even Number Matrix:")
print(even_matrix)
```

```python
# Extract third column
third_column = even_matrix[:, 2]
print("\nThird Column:")
print(third_column)
```

```python
# Set fourth row to 1,2,3,4,5
even_matrix[3, :] = [1, 2, 3, 4, 5]
print("\nModified Matrix:")
print(even_matrix)
```

```python
# 5. Consider two vectors
names = np.array(["Roxana", "Statira", "Roxana", "Statira", "Roxana"])
score = np.array([126, 115, 130, 141, 132])
```

```python
import pandas as pd
```

```python
# 1. Read the dataset into a pandas DataFrame and display the first 5 rows
print("Answer1:")

df = pd.read_csv('CarPrice_project.csv')
df.head()
```

```python
# 2. Slice the DataFrame to create a new DataFrame containing only the columns CarName, fueltype, carbody, and price. Display the first 10 rows of the new DataFrame.
selected_columns = ['CarName', 'fueltype', 'carbody', 'price']
df_selected = df[selected_columns]
df_selected.head(10)
```

```python
# 3. Filter the DataFrame to include only cars with fueltype as 'gas'. Display the first 20 rows of the filtered DataFrame.
df_gas = df[df['fueltype'] == 'gas']
df_gas.head(20)
```

```python
# 4. Slice the DataFrame to include only cars with carlength greater than 180 and fueltype as 'diesel'. Display the first 5 rows of the sliced DataFrame.
df_diesel_large = df[(df['carlength'] > 180) & (df['fueltype'] == 'diesel')]
df_diesel_large.head()
```

```python
# 5. Count the number of cars for each fueltype. Display the result.
fueltype_counts = df['fueltype'].value_counts()
fueltype_counts
```

```python
# 6. Create a new column car_type which represents the size of the car based on the carlength.
def classify_car_length(length):
    if length < 150:
        return 'short_length_cars'
    elif 150 <= length <= 170:
        return 'mid_length_cars'
    else:
        return 'long_length_cars'

df['car_type'] = df['carlength'].apply(classify_car_length)
df['car_type'].value_counts()
```

```python
# 7. Remove the column symboling from the DataFrame as an inplace operation. Display the first 5 rows of the DataFrame after removal.
df.drop(columns=['symboling'], inplace=True)
df.head()
```

```python
# 8. Group by carbody and calculate the mean price for each group.
mean_price_by_carbody = df.groupby('carbody')['price'].mean()
mean_price_by_carbody
```

```python
# 9. Create a series of 4 capital cities where the index is the name of corresponding country.
capitals = pd.Series({"India": "New Delhi", "USA": "Washington D.C.", "France": "Paris", "Japan": "Tokyo"})
capitals
```

# Matplotlib and Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("CarPrice_project.csv")
```

```python
# 1. Plot y=sin(x) and y=cos(x) on the same graph for x in [0, 2π].
x = np.linspace(0, 2 * np.pi, 100)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.legend()
plt.title("Sine and Cosine Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# 2. Create a heatmap to show the correlation matrix of numerical columns in the CarPrice_project.csv dataset.
numerical_df = df.select_dtypes(include=[np.number])  # Select only numerical columns
plt.figure(figsize=(10, 6))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 3. Create a combination plot:
# ● Use a bar plot to show the average price for each carbody.
# ● Overlay a line plot to show the count of cars for each carbody.
avg_price = df.groupby('carbody')['price'].mean()
car_counts = df['carbody'].value_counts()

fig, ax1 = plt.subplots(figsize=(10, 6))
ax2 = ax1.twinx()
ax1.bar(avg_price.index, avg_price, color='b', alpha=0.6, label='Avg Price')
ax2.plot(car_counts.index, car_counts, color='r', marker='o', label='Car Count')
ax1.set_xlabel("Car Body")
ax1.set_ylabel("Average Price", color='b')
ax2.set_ylabel("Car Count", color='r')
plt.title("Average Price & Car Count by Car Body")
plt.show()
```

```python
# 4. Visualize the relationship between horsepower, enginesize, and price.
sns.scatterplot(data=df, x='horsepower', y='price', hue='enginesize', palette='viridis')
plt.title("Horsepower vs Price (Hue: Engine Size)")
plt.show()
```

```python
# 5. Which column has the highest variance? Why might this be?
numeric_columns = df.select_dtypes(include=[np.number])
variance = numeric_columns.var().idxmax()
variance
```

```python
# 6. What insights can you draw by comparing the max and min values of the columns?
summary = df.describe().loc[['min', 'max']]
summary
```

```python
# 7. How does the variance of price compare to that of enginesize and horsepower?
variance_comparison = df[['price', 'enginesize', 'horsepower']].var()
variance_comparison
```

```python
# 8. Create a bar plot showing the maximum and minimum values for each numerical column in the dataset.
summary.T[['min', 'max']].plot(kind='bar', figsize=(12, 6))
plt.title("Max & Min Values for Each Numerical Column")
plt.show()
```

```python
# 9. Create a pie chart showing the proportion of each carbody and drivewheel type.
# ● Which carbody type dominates the dataset?
# ● Which drivewheel type is the least common?
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
df['carbody'].value_counts().plot.pie(autopct='%1.1f%%', ax=ax[0], title='Car Body Distribution')
df['drivewheel'].value_counts().plot.pie(autopct='%1.1f%%', ax=ax[1], title='Drive Wheel Distribution')
plt.show()
```
