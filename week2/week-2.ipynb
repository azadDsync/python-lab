#Numpy

The numpy package (module) is used in almost all numerical computation using Python. It is a package that provide high-performance vector, matrix and higher-dimensional data structures for Python. In the numpy package the terminology used for vectors, matrices and higher-dimensional data sets is array.



## Why is NumPy Faster Than Lists?
NumPy arrays are stored at one continuous place in memory unlike lists, so processes can access and manipulate them very efficiently.

This behavior is called locality of reference in computer science.

This is the main reason why NumPy is faster than lists. Also it is optimized to work with latest CPU architectures.

##Setting Values

```python

coffee.loc[1:3, "Units Sold"] = 10
```

The error occurs because you are attempting to use positional slicing (1:3) with .loc, which is meant for label-based indexing. Positional slicing like 1:3 should be used with .iloc.

1. Use .iloc for positional indexing: If you want to assign values to rows 1 to 3 based on their positions:

```python
coffee.iloc[1:3, coffee.columns.get_loc("Units Sold")] = 10
```

Explanation:

iloc[1:4] selects rows based on their positions (Python slicing is exclusive of the end index, so 1:4 covers rows 1, 2, and 3).
columns.get_loc("Units Sold") retrieves the column position for "Units Sold".

2. Reset or modify the index to handle positional slicing with .loc: If the index is not label-based but you want it to act like one, you can reset the index:

```python
coffee = coffee.reset_index(drop=True)  # Reset index to 0, 1, 2, ...
coffee.loc[1:3, "Units Sold"] = 10     # Now label-based slicing aligns with positional slicing

```

##apply()
a new column height_category in the DataFrame bios by applying a conditional logic to classify each individual's height (in centimeters) into categories (Short, Average, or Tall).

```python

bios['height_category'] = bios['height_cm'].apply(lambda x: 'Short' if x < 165 else ('Average' if x < 185 else 'Tall'))
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
        return 'Middleweight'

    else:
        return 'Heavyweight'

bios['Category'] = bios.apply(categorize_athlete, axis=1)
```

##Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(5,5), dpi=100)

# Line 1

# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b--', label='2x')

## Line 2

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!')

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Extract data
x = car['enginesize']
y = car['horsepower']
z = car['price']

# Normalize z to improve color variation
z_norm = (z - z.min()) / (z.max() - z.min())

# Create grid for contour
X, Y = np.meshgrid(np.linspace(x.min(), x.max(), 100), np.linspace(y.min(), y.max(), 100))
Z = np.interp(np.sqrt((X - x.mean())**2 + (Y - y.mean())**2), np.sort(z_norm), z_norm)

# Plot contour
plt.figure(figsize=(10, 6))
contour = plt.contourf(X, Y, Z, levels=50, cmap='viridis')  # Increased levels to 50
plt.colorbar(contour, label='Normalized Price')
plt.title('Contour Plot of Price Based on Engine Size and Horsepower')
plt.xlabel('Engine Size')
plt.ylabel('Horsepower')
plt.show()


```

##contour plot

```python
feature_x = np.arange(0, 50, 2)
feature_y = np.arange(0, 50, 3)

# Creating 2-D grid of features
[X, Y] = np.meshgrid(feature_x, feature_y)

fig, ax = plt.subplots(1, 1)

Z = np.cos(X / 2) + np.sin(Y / 4)

# plots contour lines
ax.contour(X, Y, Z)

ax.set_title('Contour Plot')
ax.set_xlabel('feature_x')
ax.set_ylabel('feature_y')

plt.show()
```

Plotting of contour using contourf() which plots filled contours.

```python
feature_x = np.linspace(-5.0, 3.0, 70)
feature_y = np.linspace(-5.0, 3.0, 70)

# Creating 2-D grid of features
[X, Y] = np.meshgrid(feature_x, feature_y)

fig, ax = plt.subplots(1, 1)

Z = X ** 2 + Y ** 2

# plots filled contour plot
ax.contourf(X, Y, Z)

ax.set_title('Filled Contour Plot')
ax.set_xlabel('feature_x')
ax.set_ylabel('feature_y')

plt.show()
```
